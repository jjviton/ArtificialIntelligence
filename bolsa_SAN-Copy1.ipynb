## .- Media Ultimas sesiones (SMA)

### Media Movil Simple.
_(Periodo 1 + periodo 2 + periodo 3 + periodo 4) / 4 = X_

```python
periodo =150               # Periodo para el que calculamos la Media.
totalSesiones=numeroFilas-periodo
aux=0

for y in range(totalSesiones):
    for x in range(periodo):    # Calculo para 150 sesiones.
                #print ( "valor =" ,npDatos[(y+periodo-x),4])
                #print ( "suma= ", aux)
        aux = aux + npDatos[(y+periodo-x),4]
    aux = aux/periodo
                #print(aux)  
    npDatos[(y+periodo),8]=aux
    
# Ojo: Para las primeras sesiones no puedo calcular la Media...
    
```

```python
## Calulamos la pendiente   m= IncrementoY / incrementoX
slopeRange = 4
# cojo un rango de slopeSesion para calcualr la pendiente  
aux=0
for y in range(periodo+5, numeroDatos):    
    aux = 100* (npDatos[y,8]  -  npDatos[(y-slopeRange),8])/(slopeRange)
    npDatos[y,9]=aux
    print(aux)
```

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

colors = ['teal', 'yellowgreen', 'gold']
lw = 2
fig, ax2 = plt.subplots()

## PREPARAMOS LOS DATOS DE LOS MAX Y MIN
X6 = np.zeros(numeroDatos,dtype = 'float')
#np.zeros((5,), dtype=int)
Y6 = np.zeros(numeroDatos,dtype = 'float')
X7 = np.zeros(numeroDatos,dtype = 'float')

#Preparo en Y los maximos y minimos y limpio los ceros
Y2 = npDatos[0:numeroDatos,11].copy() #Guardo la lista de max min y los ceros para el resto de datos

for var in range(0,numeroDatos):
    if(npDatos[var,11] != 0 and
       npDatos[var,11] >  0):   #MAXIMOS!!
        X6[var]=var        
        Y6[var]=npDatos[var,11]        
        #print("x=",X6[var],"y=",Y6[var], "var =",var )

#Limpio los ceros
X6 = X6[X6 != 0]                      #Limpio los ceros
Y6 = Y6[Y6 != 0]  

X6=np.reshape(X6,(-1, 1))             #transformo en columnas
Y6=np.reshape(Y6, (-1, 1))

"""print(Y6.shape[0])
print(X6.shape[1])
print("x=",X6[10:20,],"y=",Y6[10:20,] )"""

aux3=Y6.shape[0]
ax2.scatter(X6[10:aux3], Y6[10:aux3],c='red',label="training points")


# CALCULOS DE LA FUNCION POLONOMICA QUE SIGUE LA CURVA. 
# Por el momento no slip de datos, ni se como proyectar a futuro.

for count, degree in enumerate([2,4,6]):
    model = make_pipeline(PolynomialFeatures(degree), Ridge())
    model.fit(X6, Y6)
    y_plot = model.predict(X6)
    plt.plot(X6, y_plot, color=colors[count], linewidth=lw,
             label="degree %d" % degree)

plt.legend(loc='upper right')
fig.tight_layout()
plt.show()
```
